```python
import numpy as np
import matplotlib.pyplot as plt

hours = np.arange(10,24)
days = np.arange(7)

day_grid, hour_grid = np.meshgrid(days, hours, indexing="ij")

fig, ax = plt.subplots(figsize=(10,4))

ax.scatter(
    hour_grid.flatten(),
    day_grid.flatten()
)

ax.set_xticks(hours)
ax.set_yticks(days)
ax.set_xlabel("Hour")
ax.set_ylabel("Day")
ax.set_title("Day-Hour Time Slots")

ax.grid(True)

plt.show()
```
